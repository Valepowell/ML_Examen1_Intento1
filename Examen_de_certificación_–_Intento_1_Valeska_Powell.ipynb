```python
# Función para entrenar y evaluar los modelos Regresión Logística y K-Nearest Neighbors (KNN)

def entrenar_modelos(X_train, X_test, y_train, y_test, wine_type):
    # Regresión Logística
    log_reg = LogisticRegression(max_iter=1000)
    log_reg.fit(X_train, y_train)
    y_pred_lr = log_reg.predict(X_test)

    print(f"\nRegresión Logística de - {wine_type}")
    print(classification_report(y_test, y_pred_lr))
    cm = confusion_matrix(y_test, y_pred_lr)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Reds')
    plt.title(f'Matriz de Confusión')
    plt.show()

    # KNN
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)

    print(f"\nK-Nearest Neighbors- KNN de - {wine_type}")
    print(classification_report(y_test, y_pred_knn))
    cm = confusion_matrix(y_test, y_pred_knn)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
    plt.title(f'Matriz de Confusión')
    plt.show()
```

```python
print("\nMODELOS DE EVALUACION PARA VINO TINTO")
entrenar_modelos(X_train_red, X_test_red, y_train_red, y_test_red, "Vino Tinto")

print("\nMODELOS DE EVALUACION PARA VINO BLANCO")
entrenar_modelos(X_train_wt, X_test_wt, y_train_wt, y_test_wt, "Vino Blanco")
```

# Gridserch

```python

def optimizar_modelos(X_train, y_train):
    # Regresión Logística parametros
    log_reg = LogisticRegression(max_iter=1000)
    log_reg_params = {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['liblinear', 'lbfgs']
    }

    grid_log = GridSearchCV(log_reg, log_reg_params, cv=5, scoring='accuracy')
    grid_log.fit(X_train, y_train)
    best_log = grid_log.best_estimator_

    # KNN parametros
    knn = KNeighborsClassifier()
    knn_params = {
        'n_neighbors': [3, 5, 7, 9]
    }

    grid_knn = GridSearchCV(knn, knn_params, cv=5, scoring='accuracy')
    grid_knn.fit(X_train, y_train)
    best_knn = grid_knn.best_estimator_

    print("Mejores Parametros Regresión Logistica:", grid_log.best_params_)
    print("Mejores Parametros KNN :               ", grid_knn.best_params_)

    return best_log, best_knn
```

```python
def evaluar_modelos(best_log, best_knn, X_test, y_test, name):
    print(f"\nEvaluación Modelos Optimizado — {name}")

    # Regresión Logística
    y_pred_log = best_log.predict(X_test)
    print("\nRegresion Logistica Optimizada")
    print(classification_report(y_test, y_pred_log))
    cm_log = confusion_matrix(y_test, y_pred_log)
    sns.heatmap(cm_log, annot=True, fmt='d', cmap='Reds')
    plt.title(f'Matriz de Confusión Regresion Logistica')
    plt.show()

    # KNN
    y_pred_knn = best_knn.predict(X_test)
    print("\nKNN Optimizado")
    print(classification_report(y_test, y_pred_knn))
    cm_knn = confusion_matrix(y_test, y_pred_knn)
    sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Greens')
    plt.title(f'Matriz de Confusión KNN')
    plt.show()
```

```python
print("\n GridSearch Optimizando — Vino Tinto")
best_log_red, best_knn_red = optimizar_modelos(X_train_red, y_train_red)
evaluar_modelos(best_log_red, best_knn_red, X_test_red, y_test_red, "Vino Tinto")


print("\n GridSearch Optimizando — Vino Blanco")
best_log_wt, best_knn_wt = optimizar_modelos(X_train_wt, y_train_wt)
evaluar_modelos(best_log_wt, best_knn_wt, X_test_wt, y_test_wt, "Vino Blanco")
```

# Analisis Combinado de ambas DF

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# Añadir columna 'type'
df_red['type'] = 'red'
df_white['type'] = 'white'

# Unir ambos df
df_rw = pd.concat([df_red, df_white], ignore_index=True)

X_combined = df_rw.drop(columns=['quality'])
y_combined = df_rw['quality']

# Identificar columnas numéricas y categóricas
num_features_comb = X_combined.select_dtypes(include=['float64', 'int64']).columns
cat_features_comb = X_combined.select_dtypes(include=['object']).columns

# Crear preprocesador usando ColumnTransformer
preprocessor_comb = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features_comb),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_features_comb)
    ], remainder='passthrough'
)

# Aplicar preprocesamiento
X_processed_comb = preprocessor_comb.fit_transform(X_combined)


X_train_comb, X_test_comb, y_train_comb, y_test_comb = train_test_split(
    X_processed_comb, y_combined, test_size=0.2, random_state=42
)

print("\nMODELOS — Dataset Combinado")

# Regresión Logística combinado
log_reg_comb = LogisticRegression(max_iter=1000)
log_reg_comb.fit(X_train_comb, y_train_comb)
y_pred_lr_comb = log_reg_comb.predict(X_test_comb)

print("\nRegresión Logística")
print(classification_report(y_test_comb, y_pred_lr_comb))
cm_lr = confusion_matrix(y_test_comb, y_pred_lr_comb)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues')
plt.title(f'Matriz de Confusión KNN')
plt.show()

# KNN combinado
knn_comb = KNeighborsClassifier(n_neighbors=5)
knn_comb.fit(X_train_comb, y_train_comb)
y_pred_knn_comb = knn_comb.predict(X_test_comb)
print("\nKNN")
print(classification_report(y_test_comb, y_pred_knn_comb))
confusion_matrix(y_test_comb, y_pred_lr_comb)
cm_knn = confusion_matrix(y_test_comb, y_pred_lr_comb)
sns.heatmap(cm_knn, annot=True, fmt='d', cmap='Greens')
plt.title(f'Matriz de Confusión KNN')
plt.show()
```
